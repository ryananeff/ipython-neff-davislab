```python
def weighted_moving_average(x,weights,width=5):
    y = np.zeros(len(x))
    pbar = ProgressBar(len(x))
    for pos in range(0, len(x)):
        pbar.animate()
        #We're going to weight with values +/- the width
        # make sure not to wrap around (which would be bad)
        if pos-width < 0:
            start = 0
        else:
            start = pos-width
        window = x[start:pos+width] 
        weights_window = [abs(i) for i in weights[start:pos+width]]
        if sum(weights_window) == 0: # can't be normalized
            avg = np.mean(window)
        else:
            avg = np.average(window, weights=weights_window)
        y[pos] = avg
    return y
```

```python
'''

compare_alignments
Ryan Neff
7/22/2014

Usage: Compares detailed alignment statistics between two files. Should be a comparison of 
alignments between the same individual.

Inputs:
    orig_bam (string)
        The filename of a bam file aligned to the original (e.g. common) reference genome. 
    final_bam (string)
        The filename of a bam file aligned to the final (e.g. pop-specific) reference 
        genome. 
    liftover (string)
        The filename of a liftover chain file generated during the orig->final reference 
        creation process.
    out_file (string)
        (optional) The filename prefix of where the comparison statistics should be stored. 
        This defaults to "comp_aln-{datetime}".
    sample_interval (integer)
        (optional) Specifies how many bases we should sample. Defaults to 1 per 1000 bp intervals.
        Ideally, this should be set to read_length/2 (Nyquist rate?)
    chrom (string)
        (optional) Specify a chromosome to work on. Defaults to the entire BAM file. This is useful
        for parallel map/reduce functionality.
    tmp (string)
        (optional) Location of temporary files. Only used when a value is provided to chrom. Should specify a folder.

Outputs:
    {out_file}.arr
        A listing of the collected metrics by sampling location in each BAM file.
        
'''

# required imports - additional to ipy_setup.py
import pysam
import subprocess

def compare_alignments(orig_bam, final_bam, liftover, out_file=None, sample_interval=1000, 
                       metrics=None, chrom=None, vcf_file=None):
    
    # check input parameters - set those which are optional
    # I'm doing this here to make things easier to read
    if out_file==None:
        out_file = "comp_aln" + time.strftime("%m%d%y_%H%M%S")
    if metrics==None:
        metrics = [{'stepper':'all','mask':2}]        # this filters for secondary/non-primary alignments
        metrics.append({'stepper':'all','mask':3340}) # this filters for properly mapped reads
        
    # open BAM files given in input parameters
    assert pysam.Samfile(orig_bam, 'rb'), 'ERROR: Cannot open orig_bam file.'
    assert pysam.Samfile(final_bam, 'rb'), 'ERROR: Cannot open final_bam file.'
    orig_fp = pysam.Samfile(orig_bam, 'rb')
    final_fp = pysam.Samfile(final_bam, 'rb')
    
    # open the output file
    if chrom==None:
        out_fp = open(out_file + '.arr', 'wb')
    else:
        out_fp = open(out_file + "-" + chrom + ".arr", "wb")
    
    # if we are only looking at a particular chromosome, let's get rid of the extra liftover table portions
    # try to do this before we load this into memory
    if chrom != None:
        liftchrom = liftover + '.' + chrom
        subprocess.check_output('cat ' + liftover + ' | grep "' + chrom + '" > ' + liftchrom , shell=True)
        lifttable = pd.read_csv(liftchrom, header=None, sep='\t')
    else:
        lifttable = pd.read_csv(liftover, header=None, sep='\t')
        
    # for fast retrieval
    lifttable = lifttable.set_index(lifttable[2])
    
    # some more things to prep for traversal
    if chrom != None:
        chrom = [chrom]
    else:
        chrom = sorted(set(lifttable.index))
    
    # okay, we're ready to do the comparison - we're going to use a ProgressBar
    
    for cur_chrom in chrom:
        out_arr = [] # this will hold the values until we're ready to print
        end_length = lifttable.loc[cur_chrom].iloc[0][3] # this is the length of the original BAM file
        pbar = ProgressBar(len(range(0, end_length, sample_interval)))
        lift_til = 0
        delta = 0
        for pos in range(1, end_length, sample_interval):
            pbar.animate()
            fail = False
            orig_result = get_pileup_statistics(orig_fp, cur_chrom, pos) # get the result from the original BAM
            if pos > lift_til:
                try:
                    delta, lift_til = liftdelta(pos, lifttable, cur_chrom) # converts position into alt ref pos
                except:
                    fail = True
            if fail==False:
                newpos = pos+delta
                alt_result = get_pileup_statistics(final_fp, cur_chrom, newpos)
                final_result = []
                final_result.append([cur_chrom, pos])
                final_result.append([i for i in orig_result])
                final_result.append([i for i in alt_result])
                final_result = [item for sublist in final_result for item in sublist]
                out_arr.append(final_result)
        # once we're done with each chromosome, we can flush to disk (not the only solution...)
        for outstr in custom_write(out_arr):
            out_fp.write(outstr)
            
    # return if everything is happy
    return "Completed."

# used by compare_alignments - subroutine
# outputs - number of aligned reads, mean mapping quality, mean insert size, mean aligned length of reads

def get_pileup_statistics(samfile, chrom, start):
    result = (0, 0, 0, 0)
    reached = False
    for i in samfile.pileup(chrom, start, start+1):
        if i.pos == start:
            reached=True
            mapq = []
            insert = []
            aligned_score = []
            for p in i.pileups:
                mapq.append(p.alignment.mapq)
                insert.append(abs(p.alignment.tlen))
                # average alignment score
                tags = dict()
                for b in p.alignment.tags:
                    tags[b[0]] = b[1]
                aligned_score.append(tags['NM'])
            break
    if reached==True:
        result = (i.n, np.mean(mapq), np.mean(insert), np.mean(aligned_score))
    return result

def liftdelta(pos, liftover, chrom): # going in the hg19 --> AFRG1 direction
    match = liftover[(liftover.index == chrom) & (liftover[5] <= pos) & (liftover[6] > pos)].irow(0)
    return match[10] - match[5], match[6]

def custom_write(out_arr):
    for i in out_arr:
        result = ''
        for s in i:
            result += str(s) + "\t"
        result += "\n"
        yield result
```

```python
'''

compare_reads_diff
Ryan Neff
7/22/2014

Usage: Goes through all of the reads and checks to see if they have moved between references.
This also tells you where they moved to. Outputs a number of statistics. O

Inputs:
    orig_bam (string)
        The filename of a bam file aligned to the original (e.g. common) reference genome. 
    final_bam (string)
        The filename of a bam file aligned to the final (e.g. pop-specific) reference 
        genome. 
    liftover (string)
        The filename of a liftover chain file generated during the orig->final reference 
        creation process.
    out_file (string)
        (optional) The filename prefix of where the comparison statistics should be stored. 
        This defaults to "comp_reads-{datetime}".
    sample_interval (integer)
        (optional) Specifies how many bases we should sample. Defaults to 1 per 1000 bp intervals.
        Ideally, this should be set to read_length/2 (Nyquist rate?)
    chrom (string)
        (optional) Specify a chromosome to work on. Defaults to the entire BAM file. This is useful
        for parallel map/reduce functionality.
    reg (string)
        (optional) Specify a region on a chromosome to work on. Defaults to all regions in the BAM.

Outputs:
    {out_file}.arr
        A listing of the collected metrics by sampling location in each BAM file.
        
'''

# required imports - additional to ipy_setup.py
import pysam
import subprocess

def compare_reads_diff(orig_bam, final_bam, liftover, out_file=None, sample_interval=1000, chrom=None, reg=None):
    
    # check input parameters - set those which are optional
    # I'm doing this here to make things easier to read
    if out_file==None:
        out_file = "comp_reads" + time.strftime("%m%d%y_%H%M%S")
    metrics = [{'stepper':'all','mask':2}]        # this filters for secondary/non-primary alignments
    metrics.append({'stepper':'all','mask':3340}) # this filters for properly mapped reads
        
    # open BAM files given in input parameters
    assert pysam.Samfile(orig_bam, 'rb'), 'ERROR: Cannot open orig_bam file.'
    assert pysam.Samfile(final_bam, 'rb'), 'ERROR: Cannot open final_bam file.'
    orig_fp = pysam.Samfile(orig_bam, 'rb')
    final_fp = pysam.Samfile(final_bam, 'rb')
    
    # open the output file
    if chrom==None:
        out_fp = open(out_file + '.arr', 'wb')
    else:
        out_fp = open(out_file + "-" + chrom + ".arr", "wb")
    
    # if we are only looking at a particular chromosome, let's get rid of the extra liftover table portions
    # try to do this before we load this into memory
    if chrom != None:
        liftchrom = liftover + '.' + chrom
        subprocess.check_output('cat ' + liftover + ' | grep "' + chrom + '" > ' + liftchrom , shell=True)
        lifttable = pd.read_csv(liftchrom, header=None, sep='\t')
    else:
        lifttable = pd.read_csv(liftover, header=None, sep='\t')
        
    # for fast retrieval
    lifttable = lifttable.set_index(lifttable[2])
    
    # some more things to prep for traversal
    if chrom != None:
        chrom = [chrom]
    else:
        chrom = sorted(set(lifttable.index))
    
    # okay, we're ready to do the comparison - we're going to use a ProgressBar
    
    for cur_chrom in chrom:
        out_arr = [] # this will hold the values until we're ready to print
        end_length = lifttable.loc[cur_chrom].iloc[0][3] # this is the length of the original BAM file
        pbar = ProgressBar(len(range(0, end_length, sample_interval)))
        lift_til = 0
        delta = 0
        for pos in range(1, end_length, sample_interval):
            pbar.animate()
            fail=False
            orig_result = get_pileup_statistics(orig_fp, cur_chrom, pos) # get the result from the original BAM
            if pos > lift_til:
                try:
                    delta, lift_til = liftdelta(pos, lifttable, cur_chrom) # converts position into alt ref pos
                except:
                    fail=True
            if fail==False:
                newpos = pos+delta
                alt_result = get_pileup_statistics(final_fp, cur_chrom, newpos)
                final_result = []
                final_result.append([cur_chrom, pos])
                final_result.append([i for i in orig_result])
                final_result.append([i for i in alt_result])
                final_result = [item for sublist in final_result for item in sublist]
                out_arr.append(final_result)
        # once we're done with each chromosome, we can flush to disk (not the only solution...)
        for outstr in custom_write(out_arr):
            out_fp.write(outstr)
            
    # return if everything is happy
    return "Completed."

# used by compare_alignments - subroutine
# outputs - number of aligned reads, mean mapping quality, mean insert size, mean aligned length of reads

def get_pileup_statistics(samfile, chrom, start):
    result = (0, 0, 0, 0)
    reached = False
    for i in samfile.pileup(chrom, start, start+1):
        if i.pos == start:
            reached=True
            mapq = []
            insert = []
            aligned_len = []
            for p in i.pileups:
                mapq.append(p.alignment.mapq)
                insert.append(abs(p.alignment.tlen))
                aligned_len.append(p.alignment.alen)
            break
    if reached==True:
        result = (i.n, np.mean(mapq), np.mean(insert), np.mean(aligned_len))
    return result

def liftdelta(pos, liftover, chrom): # going in the hg19 --> AFRG1 direction
    match = liftover[(liftover.index == chrom) & (liftover[5] <= pos) & (liftover[6] > pos)].irow(0)
    return match[10] - match[5], match[6]

def custom_write(out_arr):
    for i in out_arr:
        result = ''
        for s in i:
            result += str(s) + "\t"
        result += "\n"
        yield result
```
