```python
def weighted_moving_average(x,weights,width=5):
    y = np.zeros(len(x))
    pbar = ProgressBar(len(x))
    for pos in range(0, len(x)):
        pbar.animate()
        #We're going to weight with values +/- the width
        # make sure not to wrap around (which would be bad)
        if pos-width < 0:
            start = 0
        else:
            start = pos-width
        window = x[start:pos+width] 
        weights_window = [abs(i) for i in weights[start:pos+width]]
        if sum(weights_window) == 0: # can't be normalized
            avg = np.mean(window)
        else:
            avg = np.average(window, weights=weights_window)
        y[pos] = avg
    return y
```

```python
import scipy.stats
```

```python
#highlight regions with significant differences in coverage, using a sliding window GOF test approach
def significant_regions(hg_val, afr_val, width=100):
    y = []
    pbar = ProgressBar(len(hg_val))
    for pos in range(0, len(hg_val)):
        pbar.animate()
        if pos-width < 0:
            start = 0
        else:
            start = pos-width
        hg_window = np.array(hg_val[start:pos+width].astype('float'))
        afr_window = np.array(afr_val[start:pos+width].astype('float'))
        fval, pval = scipy.stats.f_oneway(hg_window, afr_window)
        direction = np.mean(afr_window) - np.mean(hg_window)
        direction = direction/abs(direction) # positive if afr is larger, negative if hg is larger
        y.append(pval*direction)
    return y
```
