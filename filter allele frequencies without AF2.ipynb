```python
import sys, time
try:
    from IPython.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class ProgressBar:
    def __init__(self, iterations):
        self.iterations = iterations
        self.time = time.time()
        self.lasttime = time.time()
        self.count = 0
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython

    def animate_ipython(self):
        self.count += 1
        if time.time() - self.lasttime >= 0.5: # throttle to 0.5 secs
        #if self.count % 5000 == 0:
            print '\r', self,
            sys.stdout.flush()
            self.lasttime = time.time()
            self.update_iteration(self.count)

    def update_iteration(self, elapsed_iter):
        elapsed = time.time()-self.time # in seconds
        time_remaining = (self.iterations - elapsed_iter)*elapsed/float(elapsed_iter)
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)
        self.prog_bar += ' Elapsed: %s minutes\t Remaining: %s minutes' % (round(elapsed/60.,1), round(time_remaining/60.,1))

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) // 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
```

```python
# parallel implementation
def filterbygt(chr_name):
    import subprocess
    import pandas as pd
    #subprocess.check_output('/data/projects_gibbons/home/neffra/tools/vcftools/bin/vcftools --chr '+ chr_name
    #                        +' --vcf /data/projects_gibbons/home/neffra/63samples_combinedgenotyping.vcf.only_snps.VQSRed.vcf'
    #                        +' --recode --stdout | grep -v "#" > out.' + chr_name + '.recode.tmp', shell=True)
    inputarray = pd.read_csv('out.' + chr_name + '.recode.tmp', header=None, sep='\t')
    filteredvcf = []
    for a,i in inputarray[range(9,72)].iterrows():
        (win1, win2, loss) = (0,0,0)
        for s in list(i):
            if "./." in s:
                continue
            elif "0/0" in s:
                loss += 1
            elif "0/1" in s:
                win1 += 1
            elif "1/1" in s:
                win2 += 1
        if (win2*2+win1)>=(loss):
            filteredvcf.append(inputarray.ix[a])
    return filteredvcf
```
