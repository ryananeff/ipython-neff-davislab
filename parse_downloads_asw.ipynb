```python
gensize = pd.read_csv('/data/projects_gibbons/home/neffra/nodel_AFRGh38/genome_sizes.txt', sep='\t', header=None)
```

```python
outarr = []
totnum = 1
for a,i in gensize.iterrows():
    name = i[0]
    length = i[1]
    outarr.append([(name,length,totnum, totnum+length-1)])
    totnum += length
outarr = [item for sublist in outarr for item in sublist] #flatten
```

```python
outarr = pd.DataFrame(outarr)
```

```python
outarr
```

```python
import math
threads = 30
threads += 1
finallen = outarr.iloc[-1][3]
intervals = range(1,finallen, int(math.ceil(finallen/float(threads-1))))
intervals.append(finallen)
#print intervals
lastint = 1
lastname = outarr.iloc[0][0]
intpos = 1
threadcount = 1
filename = 'intervals_AFRG_n' + str(threads-1) + '_' + str(threadcount)
outfile = open(filename, 'wb')
for a,i in outarr.iterrows():
    name = i[0]
    while i[3] > intervals[intpos]:
        differ = i[1] - (i[3] - intervals[intpos])
        #print i[3] - intervals[intpos]
        outfile.write(''.join([str(z) for z in [name,':',lastint,'-',differ]]) + '\n')
        outfile.close()
        threadcount += 1
        filename = 'intervals_AFRG_n' + str(threads-1) + '_' + str(threadcount)
        outfile = open(filename, 'wb')
        lastint = differ + 1
        intpos += 1
    if i[3] <= intervals[intpos]:
        outfile.write(''.join([str(z) for z in [name,':',lastint,'-',i[1]]]) + '\n')
        lastint = 1
        continue
outfile.close()
    
```
