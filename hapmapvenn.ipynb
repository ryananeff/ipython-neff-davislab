```python
import math
from matplotlib_venn import venn2, venn3

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 20}
plt.rc('font', **font)

plt.figure(figsize=(10,10))
e = venn3(subsets={'100':math.log(19442), '110':math.log(1955), 
                   '111':math.log(1514365), '010':math.log(269780), 
                   '011':math.log(20478015), 
                   '001':math.log(197188), '101':math.log(929)}, normalize_to=500, 
              set_labels=("Hapmap\nVariants", "Hg38\nAlignment", "AFRG1\nAlignment"))
e.get_label_by_id('100').set_text('19,442')
e.get_label_by_id('010').set_text('269,780')
e.get_label_by_id('001').set_text('')
e.get_label_by_id('110').set_text('1,955')
e.get_label_by_id('101').set_text('929')
e.get_label_by_id('111').set_text('1,514,365')
e.get_label_by_id('011').set_text('20,478,015')
plt.annotate('197,188', xy=e.get_label_by_id('001').get_position() + np.array([3,-.4]), xytext=(0,0),
            ha='center', textcoords='offset points')
plt.show()
```
