```python
# list of files to lift
files_to_process = subprocess.check_output('ls ~/1000g/*ASW*raw.vcf', shell=True).split('\n')[0:-1]
print len(files_to_process)
print files_to_process
```

```python
'''

liftoverVCF

takes in a VCF file and a liftover array variable and lifts the coordinates in the VCF to a different reference genome. 
This uses the liftover array generated by reference generation. Works in the direction of AFRG1 --> hg19

'''

def liftoverVCF(file_name, liftover_array):
    import HTSeq as ht
    liftover = liftover_array.copy()
    vcfr = ht.VCF_Reader(file_name)
    vcfr.parse_meta()
    vcfr.make_info_dict()
    newvcf = open(file_name + "_lifted.m1.id.vcf", 'w')
    for v in vcfr:
        try:
            pos = v.pos.pos
            match = liftover[(liftover.index == v.chrom) & (liftover[10] <= pos) & (liftover[11] > pos)].irow(0)
            newvcf.write(v.to_line().replace(str(pos),str(pos + match[5] - match[10])))
        except:
            pass
    newvcf.close()
    return "Completed"
```

```python
# parallel implementation of the liftover script
balanced.block=False
pid_array = []
for file_name in files_to_process:
    pid_array.append(balanced.apply_async(liftoverVCF, file_name, liftover))
    print file_name + " sent to balanced procs."
```

```python
'''

getUnique

given a hg19 vcf and an afrg1 vcf, lift variants in one set but not another

'''

def getUnique(hg19_file, asw_file):
    import pandas as pd
    complete = pd.read_csv('../ASW_complete_no_header.vcf', sep='\t', header=None)
    hg19 = pd.read_csv(hg19_file, sep="\t", header=None)
    asw = pd.read_csv(asw_file, sep="\t", header=None)
    asw = asw.set_index(asw[0] + "_" + asw[1].map(str))
    hg19 = hg19.set_index(hg19[0] + "_" + hg19[1].map(str))
    asw_idx = set(asw.index)
    hg19_idx = set(hg19.index)
    test = asw_idx.difference(hg19_idx)
    vcf_idx = set(complete[0] + "_" + complete[1].map(str))
    test2 = test.difference(vcf_idx)
    test2_list = list(test2)
    test2_list.sort()
    well = asw[asw.index.isin(test2_list)]
    well.to_csv(asw_file + 'only_sites.vcf', sep="\t", index=False, header=False)
    return None
```

```python
'''

addVCF

combines all of the variants across a certain set of VCF files. 
This is not like the GATK implementation in that it keeps all of the extra information in each VCF file
in the final output. 

WARNING: very large files

Parallel implementation

'''

from multiprocessing import Process, Manager

def addVCF(a,i):
    file_arr = pd.read_csv(i, header=None, sep='\t')
    for b in file_arr.iterrows():
        key = ''.join([b[1][0], '\t', str(b[1][1])])
        try:
            a[key].append(list(b[1][2:10]))
        except KeyError:
            a[key] = list(b[1][2:10])

if __name__ == '__main__':
    manager = Manager()

    a = manager.dict()
    
    final_files = subprocess.check_output("ls ~/1000g/*hg19*r.vcf", shell=True).split('\n')[0:-1]
    pid_array = []
    for f in final_files:
        pid_array.append(Process(target=addVCF, args=(a,f)))
        
    for c in range(0,len(final_files)):
        pid_array[c].start()
        
    for c in range(0,len(final_files)):
        pid_array[c].join()

    with open('hg19_combined.vcf', 'w') as fp: 
        json.dump(final_arr, fp)
        
```

```python
def hist_by_chromosome(chrom, pos, centro):
    xs = []
    split_vals = []
    shade_vals =[]
    label_pos = []
    label_names = []
    y1s = []
    y2s = []
    cs = []
    last_x = 0
    names=["chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14",
           "chr15","chr16","chr17","chr18","chr19","chr20","chr21","chr22","chrX","chrY"]
    
    xy = pd.DataFrame(zip(chrom, pos), columns=['chrom', 'pos'])
    
    for name in names:
        group = xy[xy['chrom']==name]
        cstart = centro.loc[name][1] + last_x
        cend = centro.loc[name][2] + last_x
        shade_vals.append([cstart, cend])
        xs.extend(group['pos'] + last_x)
        
        # clean up
        final_val = max(group['pos'] + last_x)
        split_vals.append(final_val)
        label_pos.append((last_x+final_val)/2)
        label_names.append(name)
        last_x = final_val
    return xs, split_vals, shade_vals, label_names, label_pos
```

```python
def do_hist(xs, split_vals, shade_vals, label_names, label_pos, y_label, title, ylim=100, ymin=0):
    f = plt.figure()
    ax = f.add_axes((0.1, 0.09, 0.88, 0.85))
    plt.hist(xs, bins=int(max(xs)/5000000), log=True)
    plt.axis('tight')
    ax.vlines(split_vals,ymin,ylim,alpha=0.5)
    for i in shade_vals:
        plt.axvspan(i[0], i[1], facecolor='r', alpha=0.25)
    plt.xticks(label_pos, label_names, rotation=-90, size=16)
    plt.ylabel(y_label)
    plt.ylim([ymin,ylim])
    plt.title(title)
    plt.show()
```

```python
mb = 1000000
def do_hist2(xs, ys, split_vals, shade_vals, label_names, label_pos, y_label, title, ylim=100, ymin=0):
    f = plt.figure()
    ax = f.add_axes((0.1, 0.09, 0.88, 0.85))
    for i in range(1,len(split_vals), 2):
        plt.axvspan(split_vals[i-1], split_vals[i], facecolor='k', alpha=0.05)
    #for i in shade_vals:
    #    plt.axvspan(i[0]-5*mb, i[1]+5*mb, facecolor='g', alpha=0.25)
    plt.hist(ys, bins=int(max(xs)/5000000), log=True, color='r', alpha=0.5, label='AFRG')
    plt.hist(xs, bins=int(max(xs)/5000000), log=True, color='b', alpha=0.5, label='hg38')
    plt.axis('tight')
    #plt.axvspan(0, 1, facecolor='g', alpha=0.25, label='centromere')
    plt.xticks(label_pos, label_names, rotation=-90, size=16)
    plt.ylabel(y_label)
    plt.xlabel('chromosome')
    plt.ylim([ymin,ylim])
    plt.title(title)
    remove_border()
```

```python
'''

parallelQC

takes in a set of hg19 and afrgh19 vcf pairs and calculates mapping quality statistics for different
intersections of the sites contained therein. both vcfs need to use the same coordinate system. 
also takes in the patch vcf for another set of filtering. can take a chromosome as an input argument

'''

mqarr_d = []
def parallelQC(asw_f, hg19_f, vcf_f, chr_list=chr_list):
    # imports
    global mqarr_d
    import HTSeq as ht
    import pandas as pd
    import numpy as np
    
    # open variant files, with headers
    hg19_vcf = ht.VCF_Reader(hg19_f)
    hg19_vcf.parse_meta()
    hg19_vcf.make_info_dict()
    
    asw_vcf = ht.VCF_Reader(asw_f)
    asw_vcf.parse_meta()
    asw_vcf.make_info_dict()
    
    # open variant files, without headers, and vcf file that was added
    
    asw = pd.read_csv(asw_f, sep="\t",header=None)[[0,1,3,4]] # site and alleles
    hg19 = pd.read_csv(hg19_f, sep="\t", header=None)[[0,1,3,4]] # site and alleles
    complete = pd.read_csv(vcf_f, sep='\t', header=None)[[0,1,3,4]] # site and both alleles
    
    # set index for asw for later extraction
    asw = asw.set_index(asw[0] + "_" + asw[1].map(str))
    
    # fix contig issue
    hg19 = hg19[[x in chr_list for x in hg19[0]]]
    
    # convert to sets
    asw_idx = set(asw[0] + "_" + asw[1].map(str))
    hg19_idx = set(hg19[0] + "_" + hg19[1].map(str))
    vcf_idx = set(complete[0] + "_" + complete[1].map(str))
    
    # look at changes to pop. specific ref, changebacks, etc. - make more sets!
    asw_cng = set(asw[0] + "_" + asw[1].map(str) + "_" + asw[4])
    hg19_cng = set(hg19[0] + "_" + hg19[1].map(str) + "_" + hg19[4])
    vcf_cng_ref = set(complete[0] + "_" + complete[1].map(str) + "_" + complete[3])
    vcf_cng_alt = set(complete[0] + "_" + complete[1].map(str) + "_" + complete[4])
    
    # make critical sets
    hg19_n = hg19_idx.difference(vcf_idx)
    asw_n = asw_idx.difference(vcf_idx)
    shared = hg19_idx.intersection(asw_idx)
    uniq_hg19 = hg19_idx.difference(asw)
    uniq_asw = asw_idx.difference(hg19)
    uniq_hg19_n = hg19_n.difference(asw_n)
    uniq_asw_n = asw_n.difference(hg19_n)
    
    # make additional sets - must use different tool!
    known_h = hg19_cng.intersection(vcf_cng_alt)
    known_a = asw_cng.intersection(vcf_cng_ref)
    
    # extract the sites we changed in new ref that have a different allele
    asw_vcf_inter = asw.ix[list(asw_idx.intersection(vcf_idx))]
    asw_vcf_cng = set(asw_vcf_inter[0] + "_" + asw_vcf_inter[1].map(str) + "_" + asw_vcf_inter[4])
    diff_allele = asw_vcf_cng.difference(known_a) # result!
    
    # critical setlist
    setlist = dict({'hg19_n': [hg19_n, hg19_vcf], 
                    'asw_n': [asw_n, asw_vcf], 
                    'shared_h': [shared, hg19_vcf],
                    'shared_a': [shared, asw_vcf], 
                    'uniq_hg19': [uniq_hg19, hg19_vcf], 
                    'uniq_asw': [uniq_asw, asw_vcf],
                    'uniq_hg19_n': [uniq_hg19_n, hg19_vcf], 
                    'uniq_asw_n': [uniq_asw_n, asw_vcf]})
    
    # additional setlist
    setlist_addl = dict({'known_h': [known_h, hg19_vcf],
                         'known_a': [known_a, asw_vcf],
                         'diff_allele': [diff_allele, asw_vcf]})
    
    # write all of it to a file
    with open("/home/neffra/1000g/" + str(asw_f[19:28]) + "_setlist_crit.json", "w") as fp:
        json.dump(setlist, fp)
    with open("/home/neffra/1000g/" + str(asw_f[19:28]) + "_setlist_addl.json", "w") as fp:
        json.dump(setlist_addl, fp)
    
    # calculate mapping quality - critical sets
    qc_dict = dict()
    for s in setlist:
        mqarr = []
        svcf = setlist[s][1]
        sset = setlist[s][0]
        idt = svcf.infodict
        for i in svcf:
            chk = ''.join([str(i.chrom),"_",str(i.pos.pos)])
            if chk in sset:
                i.unpack_info(idt)
                mq = float(i.info['MQ'])
                if type(mq) == type(list()):
                    mq = mq[0] 
                mqarr.append(mq)
        try:
            qc_dict[s] = dict({'len': len(sset), 'med': np.median(mqarr), 'mean': np.mean(mqarr), 'std': np.std(mqarr)})
        except:
            mqarr_d = mqarr
            return 1
    
    # calculate mapping quality - additional sets
    for s in setlist_addl:
        mqarr = []
        svcf = setlist_addl[s][1]
        sset = setlist_addl[s][0]
        idt = svcf.infodict
        for i in svcf:
            chk = ''.join([str(i.chrom),"_",str(i.pos.pos),"_",str(i.alt[0])])
            if chk in sset:
                i.unpack_info(idt)
                mq = float(i.info['MQ'])
                if type(mq) == type(list()):
                    mq = mq[0] 
                mqarr.append(mq)
        try:
            qc_dict[s] = dict({'len': len(sset), 'med': np.median(mqarr), 'mean': np.mean(mqarr), 'std': np.std(mqarr)})
        except:
            mqarr_d = mqarr
            return 2
    with open("/home/neffra/1000g/" + str(asw_f[19:28]) + "_vcfsets.json", "w") as fp:
        json.dump(qc_dict, fp)
    return qc_dict
```
